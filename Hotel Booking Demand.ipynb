**Metric Evaluation**

1. **True Positive (TP):** Ini adalah kasus di mana model memprediksi bahwa pemesanan akan dibatalkan (kelas positif), dan prediksi tersebut benar sesuai dengan kenyataan.

2. **True Negative (TN):** Ini adalah kasus di mana model memprediksi bahwa pemesanan tidak akan dibatalkan (kelas negatif), dan prediksi tersebut benar sesuai dengan kenyataan.

3. **False Positive (FP):** Ini adalah kasus di mana model memprediksi bahwa pemesanan akan dibatalkan (kelas positif), tetapi sebenarnya pemesanan tersebut tidak dibatalkan (kelas negatif).

4. **False Negative (FN):** Ini adalah kasus di mana model memprediksi bahwa pemesanan tidak akan dibatalkan (kelas negatif), tetapi sebenarnya pemesanan tersebut dibatalkan (kelas positif).

Type I Error (False Positive):
Dalam konteks prediksi pembatalan pemesanan, ini berarti kita salah memprediksi bahwa suatu pemesanan akan dibatalkan (positif), padahal sebenarnya pemesanan tersebut tidak dibatalkan (negatif).

Tipe II Error (False Negative):
Dalam konteks prediksi pembatalan pemesanan, ini berarti kita salah memprediksi bahwa suatu pemesanan tidak akan dibatalkan (negatif), padahal sebenarnya pemesanan tersebut dibatalkan (positif).

Dalam konteks prediksi pembatalan pemesanan, mengurangi kedua jenis kesalahan ini penting untuk memastikan bahwa model memberikan prediksi yang akurat dan bermanfaat bagi manajemen hotel.

Jadi nanti metric utama yang akan kita gunakan adalah roc_auc, karena jika diperhatikan dataset yang dimiliki memiliki imbalanced data. Model ini tidak akan ada kebeperpihakan dan tidak mengutamakan FN dan FP untuk melihat evaluasi model secara holistik.

Sebelum pemodelan, data anomali akan ditangani menggunakan metode Hal ini dikarenakan:
- Distribusi data tidak normal
- Beberapa kolom memiliki data yang terpusat di sekitar satu nilai.

Berdasarkan penelusuran tersebut maka kita akan menggunakan  Metode `Interquartile Range (IQR)`.[Sumber](https://medium.com/@ayeshasidhikha188/unveiling-outliers-exploring-z-score-and-iqr-methods-for-boxplots-67b0f3086720)

```python
# Melihat lower bound, upper bound, outlier, dan persentase outlier
features = kolom_num
for i in range(0, len(features)):
    Q1 = df_model_city[features[i]].quantile(0.25)
    Q3 = df_model_city[features[i]].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound =  Q1 - (1.5 * IQR)
    upper_bound =  Q3 + (1.5 * IQR)
    condition = (df_model_city[features[i]]< lower_bound) | (df_model_city[features[i]]> upper_bound)
    outlier = df_model_city.loc[condition, features[i]].count()
    PersentaseOutlier = outlier/df_model_city[features[i]].count()*100
    print(f'''Lower Bound {features[i]} adalah {round(lower_bound,2)}, dan Upper Bound {features[i]} adalah {round(upper_bound,2)} , terdapat {outlier} data yang menyimpang dari data biasanya, yaitu sebanyak {round(PersentaseOutlier,2)}%''')
    '/n'
```

- Pada variabel `children`,`babies`, `previous_cancellations`,`previous_bookings_not_canceled`,`booking_changes`,``days_in_waiting_list`,`required_car_parking_spaces` memiliki `Lower Bound` dan `Upper Bound` 0.0 dan variabel `adults` memiliki `Lower Bound` dan `Upper Bound` yang sama yaitu 2.0 serta variabel `agent` memiliki `Lower Bound` dan `Upper Bound` yang sama yaitu 9.0 menandakan distribusi value pada variabel tersebut sangat terkonsentrasi pada nilai 0.0, 2.0 dan 9.0 sehingga variabel ini akan diabaikan tindak lanjut outliersnya karena hal ini tidak dapat dijadikan acuan.

- Pada variabel `lead_time`,`stays_in_weekend_nights`, `stays_in_week_nights`,`total_of_special_requests` terdapat outliers tetapi akan diabaikan karena dianggap wajar secara konteks memiliki variance yang tinggi.


- Outliers yang terdapat pada variabel `adr` akan ditindaklanjuti dengan cara dihapus dengan pertimbangan memiliki persentase kecil.

### **5.1 Feature Selection**  
Feature selection adalah suatu metode analisa data yang bertujuan untuk memilih fitur yang berpengaruh (fitur optimal) dan mengesampingkan fitur yang tidak berpengaruh. `Teknik Correlation Based` yang populer digunakan untuk melakukan seleksi terhadap fitur yang paling relevan dalam data set adalah `teknik Correlation Feature Selection (CFS)`. Teknik ini menghitung korelasi antara masing-masing atribut dan variabel output, lalu memilih atribut yang mempunyai nilai korelasi menengah ke atas (mendekati 1) dan membuang atribut yang memiliki nilai korelasi rendah (mendekati 0).  

Berdasarkan grafik korelasi di atas, terlihat sangat jelas pola hubungan variabel independen mana yang sangat kuat terhadap variabel target. Dari total variabel atau features awal yang dimiliki sebesar 35 akan hanya digunakan sebanyak 21.  

[Sumber 15](https://lms-paralel.esaunggul.ac.id/pluginfile.php?file=%2F313706%2Fmod_resource%2Fcontent%2F2%2F11_7705_%20MIK620_062018_pdf.pdf), [Sumber 16](https://jsi.stikom-bali.ac.id/index.php/jsi/article/download/211/170)

Nilai dari rho juga bisa di interpretasikan bagaimana makna dari nilai tersebut, dengan tabel dibawah ini:

| rho positif | rho negatif | Kategori |
| --- | --- | --- |
| 0,9 ≤ rho < 1 | -0,9 ≤ rho < -1 | Sangat kuat |
| 0,7 ≤ rho < 0,9 | -0,7 ≤ rho < -0,9 | Kuat |
| 0,5 ≤ rho < 0,7 | -0,5 ≤ rho < -0,7 | Moderat |
| 0,3 ≤ rho < 0,5 | -0,3 ≤ rho < -0,5 | Lemah |
| 0 ≤ rho< 0,3 | -0 ≤ rho < -0,3 | Sangat Lemah |

### 5.2 Feature Encoding

Untuk unique value <=4 berikutnya akan memakai one hot encoding, dan unique value >4 akan menggunakan binary encoding
- arrival_date_month : binary_encoding
- meal : one hot encoding
- country : binary encoding
- market_segment : binary encoding
- assigned_room_type : binary encoding
- deposit_type : one hot encoding
- customer_type : one hot encoding

```python
# transformer
transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first', handle_unknown = 'ignore'), ['meal','deposit_type', 'customer_type']),
    ('binary', ce.BinaryEncoder(), ['arrival_date_month','country', 'market_segment', 'assigned_room_type']),
], remainder = 'passthrough')
```

### 5.3 Feature Scaling

Mengingat dalam pembuatan model machine learning nantinya kita juga akan mencoba menggunakan algoritma Logistic Regression dan KNN maka kita akan menerapkan scaling.
Harapannya dengan memiliki fitur pada skala yang sama maka kinerja algoritma machine learning akan meningkat karena setiap fitur dapat berkontibusi sama pada target. Jika tidak dilakukan scaling maka variabel skala besar akan mendominasi fitur skala kecil khususnya untuk algoritma yang memperhitungkan jarak seperti KNN.
Kemudian karena data pada kolom numerikal terdapat outlier maka kita menggunakan RobustScaler.

```python
# Define scaler
scaler = RobustScaler()
```

## 6. Modeling & Evaluation

### 6.1 Model Benchmarking

Disini kita akan mencoba membangun dengan beberapa model klasifikasi dan kemudian nanti akan dilihat model mana yang menghasilkan roc_auc score paling tinggi pada data training dan data test.

```python
#Insert all estimators that we can use for classification model
logreg = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
xgb = XGBClassifier()
lgbm = lgb.LGBMClassifier()
```

```python
models = [logreg,knn,dt,rf,xgb,lgbm]
score=[]
rata=[]
std=[]

for i in models:
    skfold=StratifiedKFold(n_splits=10)
    estimator=Pipeline([
        ('preprocess',transformer),
        ('scaler', scaler),
        ('model',i)])
    model_cv=cross_val_score(estimator,x_train,y_train,cv=skfold,scoring='roc_auc')
    score.append(model_cv)
    rata.append(model_cv.mean())
    std.append(model_cv.std())

pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost','LightGBM'],'mean roc_auc':rata,'sdev':std}).set_index('model').sort_values(by='mean roc_auc',ascending=False)
```

```python
models = [logreg,knn,dt,rf,xgb,lgbm]
score_roc_auc = []

def y_pred_func(i):
    estimator=Pipeline([
        ('preprocess',transformer),
        ('scaler', scaler),
        ('model',i)])
    x_train,x_test

    estimator.fit(x_train,y_train)
    return(estimator,estimator.predict(x_test),x_test)

for i,j in zip(models, ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost','LightGBM']):
    estimator,y_pred,x_test = y_pred_func(i)
    y_predict_proba = estimator.predict_proba(x_test)[:,1]
    score_roc_auc.append(roc_auc_score(y_test,y_predict_proba))
    print(j,'\n', classification_report(y_test,y_pred))

pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost','LightGBM'],
             'roc_auc score':score_roc_auc}).set_index('model').sort_values(by='roc_auc score',ascending=False)
```
